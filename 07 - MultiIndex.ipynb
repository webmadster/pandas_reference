## 07.100 The .transpose() Method and MultiIndex on Column Level

## 07.103 The .unstack() Method, Part 1

## 07.104 The .unstack() Method, Part 2

```python
world = pd.read_csv('worldstats.csv', index_col=['country', 'year'])
world.head(3)
```

```python
s = world.stack()
s.head(3)
```

```python
s.unstack(-1)
s.unstack(2)

s.unstack(-2)
s.unstack(1)

s.unstack(-3)
s.unstack(0)
```

```python
s.unstack('country')
```

## 07.105 The .unstack() Method, Part 3

```python
world = pd.read_csv('worldstats.csv', index_col=['country', 'year'])
s = world.stack()
s.head(3)
```

```python
s.unstack(level = ['year', 'country'])
```

```python
s = s.unstack('year', fill_value= 0)
```

```python
s.head(3)
```

## 07.106 The .pivot() Method

```python
sales = pd.read_csv('salesmen.csv', parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales.pivot(index="Date", columns="Salesman", values="Revenue")
```

## 07.107 The .pivot_table() Method

```python
foods = pd.read_csv('foods.csv')
foods.head(3)
```

```python
foods.pivot_table(values='Spend', index=['Gender','Item'], columns='City', aggfunc='min').head(3)
```

```python
pd.pivot_table(data = foods, values='Spend', index=['Gender','Item'], columns='City', aggfunc='min').head(3)
```
